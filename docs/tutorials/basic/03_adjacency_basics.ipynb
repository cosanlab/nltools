# Adjacency Basics

This tutorial introduces the `Adjacency` class for working with similarity and distance matrices.

```python
import numpy as np
from nltools.data import Adjacency
```

## Creating Adjacency Matrices

Adjacency objects store symmetric matrices efficiently using only the upper triangle.

```python
# Create from a symmetric matrix
n = 10  # 10x10 matrix
matrix = np.random.randn(n, n)
symmetric_matrix = (matrix + matrix.T) / 2  # Make it symmetric

adj = Adjacency(symmetric_matrix)
print(f"Matrix shape: {adj.square_shape()}")
print(f"Stored as vector of length: {len(adj.data)}")
```

```python
# Create a correlation matrix from random data
data = np.random.randn(50, n)  # 50 observations of n variables
corr_matrix = np.corrcoef(data.T)

adj_corr = Adjacency(corr_matrix, matrix_type='similarity')
print(f"Correlation matrix shape: {adj_corr.square_shape()}")
```

## Basic Operations

Adjacency supports various operations while maintaining the symmetric structure:

```python
# Arithmetic operations
adj_scaled = adj * 2
adj_shifted = adj + 1

# Statistics
print(f"Mean: {adj.mean():.3f}")
print(f"Std: {adj.std():.3f}")
```

```python
# Convert between similarity and distance
# For correlation matrices: distance = 1 - correlation
adj_dist = adj_corr.distance_to_similarity(metric='correlation')
print(f"Original type: {adj_corr.matrix_type}")
print(f"Converted type: {adj_dist.matrix_type}")
```

## Visualization

Adjacency provides built-in plotting methods:

```python
# Plot the correlation matrix
adj_corr.plot()
```

```python
# Create a more structured example for visualization
# Block correlation matrix
n_blocks = 3
block_size = 5
total_size = n_blocks * block_size

block_corr = np.eye(total_size) * 0.2  # Start with small correlations
for i in range(n_blocks):
    start = i * block_size
    end = start + block_size
    block_corr[start:end, start:end] = 0.8  # High within-block correlation
    
# Add some noise
noise = np.random.randn(total_size, total_size) * 0.1
block_corr = block_corr + (noise + noise.T) / 2
np.fill_diagonal(block_corr, 1)  # Ensure diagonal is 1

adj_blocks = Adjacency(block_corr, matrix_type='similarity')
adj_blocks.plot()
```

## Working with Multiple Adjacency Matrices

```python
# Create multiple adjacency matrices
matrices = []
for i in range(5):
    data = np.random.randn(50, 10)
    corr = np.corrcoef(data.T)
    matrices.append(corr)

# Stack them
adj_multi = Adjacency(matrices, matrix_type='similarity')
print(f"Shape with multiple matrices: {adj_multi.shape()}")
```

```python
# Compute mean adjacency
mean_adj = adj_multi.mean(axis=0)
print(f"Mean adjacency shape: {mean_adj.square_shape()}")

mean_adj.plot()
```

## Converting Formats

```python
# Convert to full square matrix
square = adj_corr.squareform()
print(f"Square matrix shape: {square.shape}")

# Access the underlying vector data
vector = adj_corr.data
print(f"Vector length: {len(vector)}")
print(f"Expected length for {n}x{n} matrix: {n*(n-1)//2}")
```

## Summary

- Adjacency efficiently stores symmetric matrices (correlation, distance, similarity)
- Automatically handles upper triangle storage for memory efficiency
- Supports arithmetic operations and statistics
- Provides conversion between similarity/distance metrics
- Built-in visualization for single or multiple matrices
- Useful for connectivity analyses, RSA, and network neuroscience
