## Intelligent Flattening Rules

Brain_Collection automatically flattens to BrainData when at least 2 out of 3 dimensions are "singleton":

A dimension is singleton when it selects exactly one item:
- Position: single integer (e.g., `0`)
- Key: single string (e.g., `"beta"`)
- Label: single integer or string (e.g., `0` or `"face"`)

**Important**: 
- Lists are NEVER singleton, even `[0]` or `["beta"]`
- We NEVER concatenate across keys - they remain separate like dictionary keys

## Summary

The Brain_Collection class provides powerful 3D indexing for neuroimaging results:

- **3D indexing**: `collection[position, key, label]` for intuitive data access
- **Intelligent flattening**: Automatically returns BrainData when appropriate
- **Label support**: String-based indexing for conditions/contrasts
- **No key concatenation**: Keys remain separate like dictionary keys
- **List control**: Use lists to prevent unwanted flattening

Brain_Collection is particularly useful for:
- Managing GLM results across subjects
- Organizing multiple analysis outputs
- Extracting specific contrasts for group analysis
- Building up results incrementally
